### IMPORT THE LIBRARIES

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### LOAD THE IMAGE 

```python
img1 = cv2.imread('C:\IoT and Computer vision\color detection\images\image4.jpg')
img2 = cv2.imread('C:\IoT and Computer vision\color detection\images\image5.jpeg')
```

```python
plt.imshow(img1)
plt.title('IMAGE 1')
plt.show()
```

```python
plt.imshow(img2)
plt.title('IMAGE 2')
plt.show()
```

### CONVERT THE IMAGE FROM BGR TO RGB

```python
Bgr_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
Bgr_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(Bgr_img1)
plt.title('RGB IMAGE 1')
plt.show()
```

```python
plt.imshow(Bgr_img2)
plt.title('RGB IMAGE 2')
plt.show()
```

### CONVERT THE IMAGE FROM RGB TO HSV 

```python
Hsv_img1 = cv2.cvtColor(Bgr_img1,cv2.COLOR_RGB2HSV)
Hsv_img2 = cv2.cvtColor(Bgr_img2,cv2.COLOR_RGB2HSV)
```

### HSV 

```python
#GREEN COLOR 
greenlower_range = np.array([33,100,100])
greenupper_range = np.array([75,255,255])
```

```python
mask_img1 = cv2.inRange(Hsv_img1,greenlower_range,greenupper_range)
mask_img2 = cv2.inRange(Hsv_img2,greenlower_range,greenupper_range)
```

```python
plt.imshow(mask_img1)
plt.title('MASK IMAGE 1')
plt.show()
```

```python
plt.imshow(mask_img2)
plt.title('MASK IMAGE 2')
plt.show()
```
